```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
r = np.arange(0.01, 1.01, 0.01)
```

```python
Ea = -1.436/r
Er = (7.32e-6)/(r**8)
En = Ea + Er
```

```python
plt.plot(r, Ea, r, Er, r, En)
plt.show()
```

```python
plt.plot(r,Ea,r,Er,r,En)
plt.xlim([0.00, 1.00])
plt.ylim([-10, 10])
plt.show()
```

```python
plt.plot(r, Ea, r, Er, r, En)
plt.xlim([0.01, 1.00])
plt.ylim([-10, 10])
plt.title("attractive, repulsive and total energy of Na+ and Cl- ion pair")
plt.xlabel('Interatomic Distance in (nm)')
plt.ylabel('Bond Energy (eV)')
plt.legend(['Ea - attractive', 'Er - repulsive','En - total'])
plt.grid()
plt.show()
```

```python
# For force
F = np.diff(En)
F = 1.60217*F
r_F = np.linspace(0.01, 1.00, len(F))
```

```python
plt.plot(r_F, F)
plt.xlim([0.01, 1.00])
plt.ylim([-0.25, 0.25])
plt.title("Net force of Na+ and Cl- ion pair")
plt.xlabel('Interatomic Distance in (nm)')
plt.ylabel('Force (nN)')
plt.grid()
plt.show()
```
